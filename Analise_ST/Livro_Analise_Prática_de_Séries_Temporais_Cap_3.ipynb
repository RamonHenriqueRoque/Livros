# Dados Estacionarios

- Usando Python

```r
import pandas as pd
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import kpss
```

```r
df= pd.read_csv("https://raw.githubusercontent.com/kennedykwangari/Time-Series-Analysis-and-Forecasting-with-Python/master/AirPassengers.csv")
df
```

*Exemplo de não estacionarios*

```r
sns.lineplot(df)
```

##Teste de hipotese: *Dickey-fuller  aumentado (ADF)*

- https://www.statsmodels.org/dev/generated/statsmodels.tsa.stattools.adfuller.html

```r
adf= adfuller(list(df["#Passengers"]), autolag= "t-stat")
```

*p-Value (Segunda linha) > 5%, portando não a evidencia de ser estacionario*

```r
adf
```

##Teste de hipotese: *kwiatkowski-phillips-schmidt-shin (KPSS)*

* Calcula o teste Kwiatkowski-Phillips-Schmidt-Shin (KPSS) para a hipótese nula de que x é nível ou tendência estacionária.

- https://www.statsmodels.org/dev/generated/statsmodels.tsa.stattools.kpss.html

*Isso significa que podemos assumir que a série temporal não é estacionária em tendência.*

0.01 < 5% (0.463)

```r
model_kpss= kpss(list(df["#Passengers"]), regression= "c")
model_kpss
```

*Isso significa que a série temporal não é estacionária em termos de tendência.*

0.1 < 5% (0.146)

```r
model_kpss= kpss(list(df["#Passengers"]), regression= "ct")
model_kpss
```

# Box Cox

- temos uma maneira de transformar a lei de potência ou qualquer distribuição não linear para normal usando uma transformação Box-Cox
- https://acervolima.com/python-transformacao-box-cox/

image.png

```r
import numpy as np
from scipy import stats

# plotting modules
import seaborn as sns
import matplotlib.pyplot as plt

# generate non-normal data (exponential)
original_data = np.random.exponential(size = 1000)

# transform training data & save lambda value
fitted_data, fitted_lambda = stats.boxcox(original_data)

# creating axes to draw plots
fig, ax = plt.subplots(1, 2)

# plotting the original data(non-normal) and
# fitted data (normal)
sns.distplot(original_data, hist = False, kde = True,
            kde_kws = {'shade': True, 'linewidth': 2},
            label = "Non-Normal", color ="green", ax = ax[0])

sns.distplot(fitted_data, hist = False, kde = True,
            kde_kws = {'shade': True, 'linewidth': 2},
            label = "Normal", color ="green", ax = ax[1])

# adding legends to the subplots
plt.legend(loc = "upper right")

# rescaling the subplots
fig.set_figheight(5)
fig.set_figwidth(10)

print(f"Lambda value used for Transformation: {fitted_lambda}")
```

# Usando funções de janela

Usando R

## Janela rolante (Rolling windows)

```r
x <- rnorm(n= 100, mean= 0, sd= 10) + 1:100
mn <- function(n) rep(1/n,n)
```

```r
plot(x, type= "l", lwd= 1)
lines(filter(x, mn(5)), col= 2, lwd= 3, lty= 2)
lines(filter(x, mn(50)), col= 3, lwd= 3, lty= 3)
```

## Janelas de expansão (expanding windows)

Janelas de expansão só fazem sentido nos casos em que voce esta estimando uma sintese estatistica que acredita ser um processo estável, em vez de  evoluir com o tempo ou osilar significamente.

Em suma, supor que eh estacionário.

```r
plot(x, type= "l", lwd= 1)
lines(cummax(x), col= 2, lwd= 3, lty= 2)
lines(cumsum(x)/1:length(x), col= 3, lwd= 3, lty= 3)
lines(cummin(x), col= 4, lwd= 3, lty= 4)
```

# Self-Correlation

A autocorrelação é uma representação matemática do grau de similaridade entre uma determinada série temporal e uma versão defasada dela mesma em intervalos de tempo sucessivos

## Função de autocorrelação
- *A autocorrelação dá uma ideia de como os pontos de dados em diferentes pontos no tempo estão linearmente relacionados entre  si em função de sua diferença de tempo*

```r
x <- 1:100
y <- sin(x * pi / 3)
plot(y, type = "b")
acf(y)
```

```r
# para calcular o ACF (função autocorrelação)
library("dplyr")
cor(y, lead(y, 1), use= "pairwise.complete.obs")
cor(y, lead(y, 2), use= "pairwise.complete.obs")
```

## Função de autocorrelação Parcial (PACF)

- *A PACF mostra quais pontos de dados são informativos e quais são os pontos harmonicos em períodos de tempo mais curtos*

```r
y <- sin(x* pi / 3)
plot(y[1:30], type= "b")
pacf(y)
```

##Exemplo

```r
x = 1:100

y1 = sin(x * pi / 3)
plot(y1, type = 'b')
acf(y1)
pacf(y1)
```

```r
y2 = sin(x * pi / 10)
plot(y2, type = 'b')
acf(y2)
pacf(y2)
```

```r
# Soma das duas variaveis
y3 = y1 + y2
plot(y3, type = 'b')
acf(y3)
pacf(y3)
```

*Aumento de ruidos*

```r
noise1 = rnorm(100, sd = 0.05)
noise2 = rnorm(100, sd = 0.05)

par(mfrow = c(3, 3))
y = sin(x * pi / 4) + noise1
plot(y, type = 'b')
acf(y)
pacf(y)

y = sin(x * pi / 10) + noise2
plot(y, type = 'b')
acf(y)
pacf(y)

y = sin(x * pi / 4) + sin(x * pi / 10) + noise1 + noise2
plot(y, type = 'b')
acf(y)
pacf(y)
```
